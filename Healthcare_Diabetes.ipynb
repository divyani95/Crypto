```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
# loading the diabetes dataset to a pandas DataFrame
from google.colab import files
uploaded = files.upload()
```

```python
# Printing the first 5 rows of the dataset
# Load the dataset into a pandas DataFrame
diabetes_dataset = pd.read_csv(list(uploaded.keys())[0])

# Print the first 5 rows
print(diabetes_dataset.head())
```

```python
# Getting the number of rows and columns in the dataset
diabetes_dataset.shape
```

```python
# Getting the statistical measures of the data
diabetes_dataset.describe()
```

```python
# Examining the class distribution
diabetes_dataset['Outcome'].value_counts()
```

```python
# Separating the data into features (X) and labels (Y)
X = diabetes_dataset.drop(columns='Outcome', axis=1)
Y = diabetes_dataset['Outcome']
```

```python
# Examining the class distribution
diabetes_dataset['Outcome'].value_counts()
```

```python
# Instantiating StandardScaler
scaler = StandardScaler()
```

```python
# Fitting the scaler to our feature data (X)
scaler.fit(X)
```

```python
# Transforming the feature data using the scaler
standardized_data = scaler.transform(X)
```

```python
# Printing the standardized feature data
print(standardized_data)
```

```python
# Updating our feature data (X) with the standardized data
X = standardized_data
```

```python
# Assigning labels (Y)
Y = diabetes_dataset['Outcome']
```

```python
# Printing the standardized features and labels
print(X)
print(Y)
```

```python
# Splitting the dataset into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
# Printing the shapes of the original, training, and testing feature sets
print(X.shape, X_train.shape, X_test.shape)
```

```python
# Instantiating the Support Vector Machine classifier
classifier = svm.SVC(kernel='linear')
```

```python
# Training the Support Vector Machine Classifier
classifier.fit(X_train, Y_train)
classifier = svm.SVC(kernel='linear')
classifier.fit(X_train, Y_train)
```

```python
# Accuracy score on the training data
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
# Printing the accuracy score on the training data
print('Accuracy score of the training data: ', training_data_accuracy)
```

```python
# Accuracy score on the test data
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
# Printing the accuracy score on the test data
print('Accuracy score of the test data: ', test_data_accuracy)
```
